```python

# GENDER-DISTRIBUTION
Gender_count=df['Gender'].value_counts()

plt.figure(figsize=(8,9))

plt.pie(
    Gender_count,
    labels=Gender_count.index,
    colors=["blue","pink"],
    autopct="%1.1f%%",
    explode=[0,0.1]
)
plt.title("GENDER DISTRIBUTION")

plt.show()
```

```python
# EDUCATION SECTOR
educationsector_count=df['EducationSector'].value_counts()

plt.figure(figsize=(10,6))

plt.bar(
    educationsector_count.index,
    educationsector_count.values,
   
    color="skyblue"
    
    
)
plt.title("EDUCATION SECTOR DISTRIBUTION")
plt.x_label=("EducationSector")
plt.xticks(rotation=90)

plt.legend()
plt.show()
```

```python
IndividualProject_count=df['IndividualProject'].value_counts()
plt.figure(figsize=(7,8))
plt.pie(
    IndividualProject_count,
    labels=IndividualProject_count.index,
    colors=["orange","purple"],
    autopct="%1.1f%%"


)
plt.title("Students with Individual Projects")
plt.show()
```

```python
Influenced_count=df['Influenced'].value_counts()
plt.figure(figsize=(6,7))
plt.pie(
    Influenced_count,
    labels=Influenced_count.index,
    colors=["silver","grey"],
    
    autopct="%1.1f%%"


)
plt.title("INFLUENCED")
plt.show()
```

```python
y_count=df['y'].value_counts()
plt.figure(figsize=(6,7))
plt.pie(
    y_count,
    labels=y_count.index,
    colors=["skyblue","blue"],
    
    autopct="%1.1f%%"


)
plt.title("Entrepreneurial Ability Distribution")
plt.show()
```

```python
city_y = pd.crosstab(df["City"], df["y"])

city_y.plot(
    kind="bar",
    figsize=(10,6),
    color=["skyblue", "pink"]
)

plt.title("Entrepreneurial Ability by City")
plt.xlabel("City")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.legend(title="Entrepreneurial Ability")

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.hist(
    df["Age"],
    bins=10,
    color="silver",
    edgecolor="black"
)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Students")

plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(
    df['Age']
    )

plt.title("Age Box Plot")
plt.ylabel("Age")

plt.show()
```
