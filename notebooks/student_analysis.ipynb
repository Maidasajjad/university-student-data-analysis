```python

# GENDER-DISTRIBUTION
Gender_count=df['Gender'].value_counts()

plt.figure(figsize=(8,9))

plt.pie(
    Gender_count,
    labels=Gender_count.index,
    colors=["blue","pink"],
    autopct="%1.1f%%",
    explode=[0,0.1]
)
plt.title("GENDER DISTRIBUTION")

plt.show()
```

Observation:
• The pie chart shows the percentage distribution of male and female students.
• It helps compare the gender ratio in the dataset.
• The larger slice represents the gender with more students.

. The tallest bar indicates the education sector with the highest number of students.

. The chart helps compare student counts across education sectors.

```python
plt.figure(figsize=(10,8))

numeric_df = df.select_dtypes(include="number")

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

The heatmap shows the correlation between numerical variables in the dataset.
Values closer to +1 indicate a strong positive relationship, meaning both variables increase together.
Values closer to -1 indicate a negative relationship, meaning one variable increases while the other decreases.
Values around 0 show a weak or no relationship between variables.
The heatmap helps identify which features have a strong relationship with each other and can be useful for data analysis and feature selection.

```python
sns.pairplot(
    data=df,
    vars=[
        "Age",
        "Perseverance",
        "Competitiveness",
        "SelfConfidence"
    ],
    hue="y"
)

plt.show()
   
```

The pair plot shows the relationships and distributions among Age, Perseverance, Competitiveness, and SelfConfidence variables.
The diagonal plots represent the distribution of each variable individually.
The scatter plots show how these features are related to each other.
It helps identify whether higher Perseverance and Competitiveness are associated with higher SelfConfidence.
The relationship between Age and other factors can also be observed through the patterns of data points.
A positive trend indicates that as one variable increases, the other variable also tends to increase.
A scattered pattern indicates a weak or no clear relationship between variables.
Overall, the pair plot helps understand patterns, correlations, and relationships among important student characteristics.

```python
sns.kdeplot(
    data=df,
    x="Age",
    fill=True
)

plt.show()
```

The KDE plot shows the distribution pattern of students' ages.
The highest peak represents the age range where most students are concentrated.
A wider curve indicates more variation in the age values.
The plot helps identify the most common age group and the overall spread of ages in the dataset.
If the curve is smooth and centered around a specific age, it shows that most students belong to that age group.
The KDE plot provides a clear view of the age distribution compared to a simple histogram.
